# Support Vector Machine

A dichotomous classifier that maximize the margine between two classes. Data usually labeled as -1 or 1. 

# Problem Formulation

## Problem Types
### Linear Separable
Easy problem, model can be trained using Gradient Descent
### Non-linear Separable
Hard probelm, model can be trained using Kernel Trick and Sequential Minimal Optimization method 
### Non-separable
Use soft margin to tolerance a small amount of error 

## Model
source:https://www.csie.ntu.edu.tw/~cjlin/papers/libsvm.pdf
### Primal Problem
Data are separated by a hyperplane. The hyperplane can be described using <w, b>, thus the primal problem of SVM is:
$$\min_{w,b,\xi}\frac{1}{2}w^{T}w+C\sum_{i=1}^{N}{\xi_{i}}$$
$$s.t.\space y_i(w\cdot K(x_i) + b) \geq 1 - \xi_i$$
$$\xi \geq 0, i = 1,..,N$$

### Dual Problem
Using Lagrange multiplers, the primal problem is transformed to a computational-easier probelm:
$$L(ω,b,\xi,\alpha,\mu) = \frac{1}{2}w^{T}w+C\sum_{i=1}^{N}{\xi_{i}} - \sum_{i=1}^{N}{\alpha_i(y_i(w\cdot K(x_i) + b) - 1 + \xi_i)} - \sum_{i=1}^{N}{\mu_i\xi_i}$$
$$\max_{\alpha,\mu}\min_{w,b,\xi}L(ω,b,\xi,\alpha,\mu)$$
$$s.t.\space \alpha,\mu\geq0$$
Equivalent to
$$\min_{\alpha}\frac{1}{2}\sum_{i=1}^{N}{\sum_{j=1}^{N}{\alpha_i\alpha_jy_iy_jK(x_i\cdot x_j)}} - \sum_{i=1}^{N}{\alpha_i}$$
$$s.t.\space \sum_{i=1}^{N}{\alpha_iy_i} = 0$$
$$0 \leq \alpha_i \leq C, i = 1,...,N$$

In this project, we will implement an SVM using both approaches and compare their performance.\

```python
d = np.genfromtxt('nonlinear_data.txt')
X,Y = d[:,:2], d[:,-1]

kernel = Kernel()

for k in kernel.family:
    svm = SVM(k, C = 20)
    svm.fit(X, Y)
    svm.error(X, Y)
    svm.plot(title = 'Kernel: '+k.__name__.title())
    print(svm.report)
```

```python
d = np.genfromtxt('linear_data.txt')
X,Y = d[:,:2], d[:,-1]

kernel = Kernel()
for k in kernel.family:
    svm = SVM(k, C = 20)
    svm.fit(X, Y)
    svm.error(X, Y)
    svm.plot(title = 'Kernel: '+k.__name__.title())
    print(svm.report)
```
