## Demo for primal SVM using QP

```python
import numpy as np
import matplotlib.pyplot as plt
from SVM_QP import SVM_QP as SVM

%load_ext autoreload
%autoreload 2
```

## Hard Margin SVM for linearly separable data

```python
data = np.loadtxt('linear_data.txt',delimiter='    ')
x = data[:,:-1]
y = data[:,-1]
model = SVM()
sol = model.fit(x,y)
```

```python
model.error(x,y)
```

```python
model.plot(x,y,title='Hard')
```

## Soft Margin SVM for non-linearly separable data

```python
data = np.loadtxt('nonlinear_data.txt',delimiter='    ')
x = data[:,:-1]
y = data[:,-1]
model = SVM()
sol = model.fit(x,y,1.0) # c = 0.5 for soft svm
```

```python
model.error(x,y)
```

```python
model.plot(x,y,title='Hard')
```
